## 1. Génération de suites

Trois façons de produire les termes d'une suite : terme général $u_n = f(n)$, relation de récurrence $u_{n+1} = f(u_n)$, et le cas classique de Fibonacci.

```python
def terme_general(f , debut , fin):
    """ renvoie la liste pour u_n = f(n) """

    return [f(n) for n in range(debut, fin + 1)]

# jeux de tests

u = lambda n : (-1) ** n / n
print(terme_general(u,1,5))
```

```python
def suite_recurrente(f,u0,N):
    """ renvoie [u0, u1, ..., uN] pour la recurrence """

    termes = [u0]
    u = u0
    for _ in range (N):
        u = f(u)
        termes.append(u)
    return termes


# jeux de tests
from math import sqrt
f = lambda x : 1 + sqrt(x)
print(suite_recurrente(f,2,6))
```

```python
def fibonacci(N):
    """ renvoie les N+1 premiers termes de la suite de Fibonacci (u0 = u1 = 1) """

    termes = [1,1]
    for _ in range(N - 1):
        termes.append(termes[-1] + termes[-2])
    return termes

print(fibonacci(10))
```

## 2. Propriétés d'une suite : majoration, minoration, bornitude, monotonie

Chaque fonction prend en entrée une liste de termes déjà calculée (via `terme_general` ou `suite_recurrente`) et teste la propriété correspondante — même logique que les exercices 1.1-1.3 du carnet.

```python
def est_majoree(termes):
    """ teste sur les termes fournis """

    return max(termes)
```

```python
def est_minoree(termes):
    """ teste sur les termes fournis """

    return min(termes)
```

```python
def est_bornee(termes):
    """ majoree et bornee """

    return max(abs(u) for u in termes)
```

```python
def est_croissante(termes):
    return all(termes[n+1] >= termes[n] for n in range(len(termes) -1))
               
```

```python
def est_decroissante(termes):
    return all(termes[n+1] <= termes[n] for n in range(len(termes) - 1))
```

```python
def est_monotone(termes):
    """ croissante ou decroissante """
    return est_croissante(termes) or est_decroissante(termes)
    
```

## 3. Vérification sur un exemple

Application des fonctions ci-dessus à $u_n = \dfrac{(-1)^n}{n}$ pour $n \in \{1, \dots, 99\}$.

```python
u = lambda n : (-1)**n / n
termes = [u(n) for n in range(1,100)]

print("majorée par :", est_majoree(termes))     # ~0.5
print("minorée par :", est_minoree(termes))     # -1.0
print("bornée, |u_n| ≤ :", est_bornee(termes))  # 1.0
print("croissante ?", est_croissante(termes))   # False
print("décroissante ?", est_decroissante(termes)) # False
print("monotone ?", est_monotone(termes))       # False
```

```python
import numpy as np

```

```python
rng = np.random.default_rng(42)
n = 4

def vecteur():
    return rng.integers(-9,10, size = n).astype(float)

def scalaire():
    return float(rng.integers(-9,10))

u , v , w = vecteur(), vecteur(),vecteur()
lam , mu = scalaire() , scalaire()

print("u =", u)
print("v =", v)
print("w =", w)
print("λ =", lam, "  µ =", mu)


                 
```

### 4.1 Axiomes de structure d'espace vectoriel — addition

Commutativité, associativité, élément neutre, symétrique (opposé).

```python
zero = np.zeros(n)

# 1. Commutativité
print("1. Commutativité   u+v = v+u        :", np.allclose(u + v, v + u))

# 2. Associativité
print("2. Associativité   u+(v+w)=(u+v)+w  :", np.allclose(u + (v + w), (u + v) + w))

# 3. Élément neutre
print("3. Neutre          u + 0 = u        :", np.allclose(u + zero, u))

# 4. Symétrique (opposé)
print("4. Symétrique      u + (-u) = 0     :", np.allclose(u + (-u), zero))

```

### 4.2 Axiomes de structure d'espace vectoriel — multiplication scalaire

$1 \cdot u = u$, associativité mixte $\lambda(\mu u) = (\lambda\mu)u$, distributivité sur les vecteurs et sur les scalaires.

```python
# 5. 1 · u = u
print("5. 1·u = u                          :", np.allclose(1 * u, u))

# 6. λ·(µ·u) = (λµ)·u
print("6. λ·(µ·u) = (λµ)·u                 :", np.allclose(lam * (mu * u), (lam * mu) * u))

# 7. Distributivité / addition des vecteurs : λ·(u+v) = λu + λv
print("7. λ·(u+v) = λu + λv                :", np.allclose(lam * (u + v), lam * u + lam * v))

# 8. Distributivité / addition des scalaires : (λ+µ)·u = λu + µu
print("8. (λ+µ)·u = λu + µu                :", np.allclose((lam + mu) * u, lam * u + mu * u))
```

### 4.3 Propriétés élémentaires déduites des axiomes

$0 \cdot u = 0_E$, $\lambda \cdot 0_E = 0_E$, $(-1) \cdot u = -u$.

```python
# Rappel : zero = np.zeros(n) est le vecteur nul 0_E
zero = np.zeros(n)

# 1. 0 · u = 0_E   (le scalaire 0 annule tout vecteur)
print("1. 0·u = 0_E              :", np.allclose(0 * u, zero))

# 2. λ · 0_E = 0_E  (multiplier le vecteur nul donne le vecteur nul)
print("2. λ·0_E = 0_E            :", np.allclose(lam * zero, zero))

# 3. (-1) · u = -u  (le scalaire -1 donne l'opposé)
print("3. (-1)·u = -u           :", np.allclose((-1) * u, -u))
```

### 4.4 Intégrité : $\lambda \cdot u = 0_E \implies \lambda = 0$ ou $u = 0_E$

Un scalaire non nul et un vecteur non nul ne peuvent jamais donner le vecteur nul par multiplication scalaire.

```python
# Sens intéressant : si λ·u = 0_E, alors forcément λ=0 OU u=0_E.
# Autrement dit : un scalaire non nul et un vecteur non nul ne peuvent
# JAMAIS donner le vecteur nul.

# Test : λ non nul, u non nul  ->  λ·u ne doit PAS être nul
lam_nonnul = 3.0
u_nonnul = vecteur()
produit = lam_nonnul * u_nonnul

print("λ =", lam_nonnul, " non nul")
print("u =", u_nonnul, " non nul")
print("λ·u =", produit)
print("λ·u est-il nul ?", np.allclose(produit, zero))   # attendu : False
```

### 4.5 Sous-espaces vectoriels : critère et exemples

Un sous-ensemble $F \subset E$ est un sous-espace vectoriel si et seulement si : (1) $0_E \in F$, (2) $F$ stable par addition, (3) $F$ stable par multiplication scalaire — condition résumée par la stabilité par combinaison linéaire $\lambda u + \mu v \in F$.

#### Contre-exemple 1 : $F_1 = \{(x,y) \mid x+y=2\}$ (droite ne passant pas par l'origine)

```python
# F1 = { (x, y) ∈ R² | x + y = 2 }
# C'est une droite... mais qui ne passe PAS par l'origine.

def dans_F1(vec):
    x, y = vec
    return np.isclose(x + y, 2)

origine = np.array([0.0, 0.0])
print("F1 contient-il 0_E ?", dans_F1(origine))   # False -> déjà mort

# Pire : F1 n'est même pas stable par addition
a = np.array([1.0, 1.0])   # 1+1 = 2  ✓
b = np.array([0.0, 2.0])   # 0+2 = 2  ✓
print("a ∈ F1 ?", dans_F1(a), " | b ∈ F1 ?", dans_F1(b))
print("a+b ∈ F1 ?", dans_F1(a + b))   # (1,3) -> 1+3=4 ≠ 2 -> False
```

#### Contre-exemple 2 : $F_2 = \{(x,y) \mid x=0 \text{ ou } y=0\}$ (réunion des deux axes)

Contient bien $0_E$, mais n'est pas stable par addition.

```python
# F2 = { (x, y) ∈ R² | x = 0 OU y = 0 }  (les deux axes)
# Contient bien 0_E, mais...

def dans_F2(vec):
    x, y = vec
    return np.isclose(x, 0) or np.isclose(y, 0)

u2 = np.array([1.0, 0.0])   # sur l'axe des x
v2 = np.array([0.0, 1.0])   # sur l'axe des y
print("u2 ∈ F2 ?", dans_F2(u2), " | v2 ∈ F2 ?", dans_F2(v2))
print("u2+v2 =", u2 + v2, " ∈ F2 ?", dans_F2(u2 + v2))   # (1,1) -> False
```

#### Exemple validé : $F = \{(x,y) \mid x+y=0\}$ (droite passant par l'origine)

Les trois critères sont vérifiés un par un.

```python
# F = { (x, y) ∈ R² | x + y = 0 }  (une droite passant par l'origine)

def dans_F(vec):
    x, y = vec
    return np.isclose(x + y, 0)

zero2 = np.array([0.0, 0.0])

# --- Condition 1 : 0_E ∈ F ? ---
print("1. 0_E ∈ F ?", dans_F(zero2))    # 0+0=0 -> True

# --- Condition 2 : stable par addition ? ---
# On fabrique deux éléments DE F au hasard.
# Astuce : un élément de F s'écrit (t, -t) pour t quelconque.
t1, t2 = 5.0, -3.0
a = np.array([t1, -t1])    # (5, -5)   ✓ dans F
b = np.array([t2, -t2])    # (-3, 3)   ✓ dans F
print("   a =", a, "∈ F ?", dans_F(a))
print("   b =", b, "∈ F ?", dans_F(b))
print("2. a+b =", a + b, "∈ F ?", dans_F(a + b))   # (2,-2) -> True

# --- Condition 3 : stable par multiplication scalaire ? ---
lam = -4.0
print("3. λ·a =", lam * a, "∈ F ?", dans_F(lam * a))   # (-20,20) -> True
```

#### Vérification globale : critère unique (combinaison linéaire) sur 1000 tirages aléatoires

```python
# Théorème 3 : F est un SEV  <=>  λu + µv ∈ F  pour tous u,v ∈ F et λ,µ ∈ K.
# On teste cette condition unique (qui résume les 3) sur plein de cas au hasard.

def element_de_F():
    # génère un élément (t, -t) donc garanti dans F
    t = float(rng.integers(-9, 10))
    return np.array([t, -t])

tout_ok = True
for _ in range(1000):
    u_, v_ = element_de_F(), element_de_F()
    lam_, mu_ = float(rng.integers(-9, 10)), float(rng.integers(-9, 10))
    combinaison = lam_ * u_ + mu_ * v_
    if not dans_F(combinaison):
        tout_ok = False
        break

print("λu + µv reste dans F sur 1000 essais ?", tout_ok)   # True
```

### 4.6 Somme et somme directe de sous-espaces vectoriels

$F+G = \{u+v \mid u \in F,\, v \in G\}$. La somme est **directe** (notée $F \oplus G$) si de plus $F \cap G = \{0_E\}$ : tout élément de $F+G$ s'écrit alors de façon **unique**.

#### Exemple 15 : $F+G$ dans $\mathbb{R}^3$ — $F$ = axe des $x$, $G$ = axe des $y$

```python
# F + G = { u + v | u ∈ F, v ∈ G }
# Exemple 15 du PDF, dans R³ :
#   F = { (x,0,0) }  (axe des x)
#   G = { (0,y,0) }  (axe des y)
# Un élément de F s'écrit (s,0,0), un élément de G s'écrit (0,t,0).

def elem_F():
    s = float(rng.integers(-9, 10))
    return np.array([s, 0.0, 0.0])

def elem_G():
    t = float(rng.integers(-9, 10))
    return np.array([0.0, t, 0.0])

# On génère des éléments de F+G et on regarde leur forme
print("Quelques éléments de F + G :")
for _ in range(5):
    somme = elem_F() + elem_G()
    print("  ", somme)

# Observation : tous de la forme (x, y, 0) -> F+G = { (x,y,z) | z = 0 }
```

#### Exemple 16 : une somme non directe — $F \cap G \neq \{0_E\}$

Deux plans de $\mathbb{R}^3$ dont l'intersection est une droite : la décomposition d'un vecteur en $F+G$ n'est plus unique.

```python
# Exemple 16 du PDF :
#   F = { (x,y,z) | x = 0 }  (plan yz)
#   G = { (x,y,z) | y = 0 }  (plan xz)
# On a bien F + G = R³, MAIS la somme n'est PAS directe.

# Preuve : l'intersection n'est pas réduite à {0}.
# F ∩ G : x=0 ET y=0  ->  { (0,0,z) } = tout l'axe z, pas seulement 0 !

# Conséquence : décomposition NON unique. Le PDF donne (1,2,3) :
w = np.array([1.0, 2.0, 3.0])

# Décomposition n°1 : (0,2,3) + (1,0,0)
d1_F = np.array([0.0, 2.0, 3.0]); d1_G = np.array([1.0, 0.0, 0.0])
# Décomposition n°2 : (0,2,0) + (1,0,3)
d2_F = np.array([0.0, 2.0, 0.0]); d2_G = np.array([1.0, 0.0, 3.0])

print("Décomp 1 :", d1_F, "+", d1_G, "=", d1_F + d1_G)
print("Décomp 2 :", d2_F, "+", d2_G, "=", d2_F + d2_G)
print("Deux décompositions différentes du MÊME w -> somme NON directe")
print("(car F ∩ G = axe z ≠ {0})")
```

#### Exemple classique de somme directe : fonctions paires $\oplus$ fonctions impaires

Toute fonction $f : \mathbb{R} \to \mathbb{R}$ se décompose de façon unique en $f = f_p + f_i$ avec $f_p$ paire et $f_i$ impaire — car $P \cap I = \{0\}$ (seule la fonction nulle est à la fois paire et impaire).

```python
def est_paire(f,essais = 200):
    xs = rng.uniform(-10,10,essais)
    return np.allclose([f(-x) for x in xs], [f(x) for x in xs])

def est_impaire(f, essais = 200):
    xs = rng.uniform(-10,10,essais)
    return np.allclose([f(-x) for x in xs], [-f(x) for x in xs])

#jeux de tests

print("cos paire ?   ", est_paire(np.cos))
print("sin impaire ? ", est_impaire(np.sin))
print("x³ impaire ?  ", est_impaire(lambda x: x**3))
print("x² paire ?    ", est_paire(lambda x: x**2))
```

```python
def paire_et_impaire(f):
    return est_paire(f) and est_impaire(f)

print("fonction nulle -> P ∩ I ?", paire_et_impaire(lambda x: 0*x))   # True
print("cos -> P ∩ I ?          ", paire_et_impaire(np.cos))          # False
```

### 5.1 Rang à partir duquel $|u_n - \ell| < \varepsilon$ (définition $\varepsilon$-$N$)

```python
def rang_n(u,ell,eps,fenetre = 50,nmax = 100000):
    """ renvoie le rang n à partir duquel |u_n - ell| < eps """

    for n in range(1,nmax):
        if all(abs(u(k) - ell) < eps for k in range(n, n + fenetre)):
            return n
    return None

# jeux de tests

u = lambda n : (2*n + 1)/(n + 2)

for eps in [0.1, 0.01, 0.001]:
    N = rang_n(u, 2, eps)
    print(f"Pour ε = {eps}, |u_n - 2| < ε pour n ≥ {N} (u_{N} = {u(N)})")
```

### 5.2 Théorème des gendarmes

```python
def encadrement_ok(u,v,w,debut,fin):
    """verifie le theoreme des gendarmes pour n debut a fin teste"""

    return all(u(n) <= v(n) <= w(n) for n in range(debut, fin + 1))

# jeux de tests
u = lambda n: 2 - 1/(1 + n + n**2)
v = lambda n: 2 + (-1)**n / (1 + n + n**2)
w = lambda n: 2 + 1/(1 + n + n**2)

print("Encadrement ok pour n=1..100 ?", encadrement_ok(u,v,w,0,1000))

for n in [1, 10, 100, 1000]:
    print(f"n={n:<5} u={u(n):.5f}  v={v(n):.5f}  w={w(n):.5f}")

```

### 5.3 Suite géométrique $u_n = a^n$

```python
def geometrique(a,N):
    """ renvoie les N+1 premiers termes de la suite géométrique u_n = a^n """

    return [a**n for n in range(N + 1)]

# jeux de tests

for a in [1,2,0.5,-1.5]:
    termes = geometrique(a, 6)
    apercu = [round(t,3) for t in termes]
    print(f"a={a:<5}  u_n = {apercu}")
```

### 6.7 PGCD — algorithme d'Euclide

```python
def pgcd(a,b):
    """renvoie le plus grand commun diviseur de a et b"""
    while b != 0:
        a, b = b, a % b
    return a 
# jeux de tests
print(pgcd(12,18))  # 6
print(pgcd(48,18)) # 6
print(pgcd(17, 5))
```

```python
def pgcd_rec(a,b):
    if b == 0:
        return a
    return pgcd_rec(b, a % b)

# jeux de tests
print(pgcd_rec(12,18))  # 6 
print(pgcd_rec(48,18))  # 6
```

```python
import math
def coprimes(a,b):
    return math.gcd(a,b) == 1

print(coprimes(12,18))  # False
print(coprimes(7,40))   # True
```

### 6.8 Algorithme d'Euclide étendu et théorème de Bézout

$d = \mathrm{pgcd}(a,b) = au + bv$.

```python
def euclide_etendu(a,b):
    """renvoie le couple (d,x,y) tel que d = pgcd(a,b) et d = ax + by"""
    if b == 0:
        return a, 1, 0
    g,u1,v1 = euclide_etendu(b, a % b)
    u = v1
    v = u1 - (a // b) * v1
    return g,u,v
# jeux de tests
g,u,v = euclide_etendu(48,18)
print(g,u,v)
print(48*u + 18*v)
```

### 6.9 Inverse modulaire

$a$ est inversible modulo $n$ ssi $\mathrm{pgcd}(a,n) = 1$ ; l'inverse se lit dans l'identité de Bézout $au+nv=1$.

```python
def inverse_modulaire(a,n):
    """renvoie l'inverse de a modulo n si pgcd(a,n) = 1, None sinon"""
    g,u,v = euclide_etendu(a,n)
    if g != 1:
        return None
    return u % n

#jeux de tests
print(inverse_modulaire(3,7))   # 5
print(inverse_modulaire(2,6))   # None  
print(inverse_modulaire(7,40))
```

### 6.11 Exponentiation rapide (par carrés successifs)

```python
def exp_modulaire(base, exp, mod):
    """Calcule base^exp mod mod par exponentiation rapide."""
    resultat = 1
    base = base % mod
    while exp > 0:
        if exp % 2 == 1:          # bit de poids faible = 1
            resultat = (resultat * base) % mod
        exp = exp // 2            # on décale
        base = (base * base) % mod # on élève au carré
    return resultat

print(exp_modulaire(7, 100, 13))   # 9
print(pow(7, 100, 13))             # 9  -> même résultat
```

### 7.1 Somme d'une série géométrique — vérification numérique de la formule

```python
def serie_directe(a,N):
    """somme en additionnant terme a termes"""
    return sum(a ** n for n in range(N + 1))

def serie_formule(a,n):
    """meme somme mais via la formule de la somme d'une suite géométrique"""

    return (1 - a **(N + 1)) / (1 - a)

a = 0.5

for N in [5, 10, 50]:
    directe = serie_directe(a, N)
    formule = serie_formule(a, N)
    print(f"N={N:<4} directe={directe:.6f}  formule={formule:.6f}")

print(f"limite théorique 1/(1-a) = {1/(1-a)}")

```

### 7.2 Suite $a^n/n!$ — vérification numérique de l'exercice 1.3 (décroissance à partir d'un certain rang)

```python
from math import factorial

def a_n_sur_factorielle(a, N):
    """renvoie le terme général a_n = n! / (n+1)!"""
    return [a**n / factorial(n) for n in range(N + 1)]

a = 100
termes = a_n_sur_factorielle(a, 250)

# jeux de tests
for n in[10,50,100,150,250]:
    print(f"n={n:<5} a^n/n! = {termes[n]:.3e}")
```

## 8. Retour à l'arithmétique : petit théorème de Fermat et indicatrice d'Euler

### 8.1 Vérification du petit théorème de Fermat

```python
def verifie_fermat(a,p):
    """verifie le petit théorème de Fermat : a^(p-1) ≡ 1 (mod p)"""
    return pow(a, p - 1, p) == 1
# jeux de tests
print(verifie_fermat(3, 7))     # True   -> 3^6 ≡ 1 (mod 7)
print(verifie_fermat(2, 13))    # True   -> 2^12 ≡ 1 (mod 13)
```

### 8.2 Indicatrice d'Euler $\varphi(n)$

```python
from math import gcd

def phi(n):
    """renvoie le nombre d'entiers positifs ≤ n et premiers avec n"""
    return sum(1 for k in range(1, n + 1) if gcd(k, n) == 1)

# jeux de tests
print(phi(1))   # 1
print(phi(2))   # 1
print(phi(3))   # 2
print(phi(4))   # 2
print(phi(5))   # 4     
```

## 9. Retour au chiffrement de César : cryptanalyse par force brute

On essaie les 26 décalages possibles pour retrouver le message en clair — la faiblesse structurelle du chiffre de César (espace de clés minuscule).

```python
message = "WKH VHFUHW LV KLGGHQ LQ WKH JDUGHQ"

def dechiffre_cesar(texte, cle):
    """Déchiffre 'texte' avec un décalage 'cle'."""
    resultat = ""
    for c in texte:
        if c.isalpha():
            pos = ord(c) - ord('A')          # 'A'->0 ... 'Z'->25
            nouvelle_pos = (pos - cle) % 26   # décalage inverse mod 26
            resultat += chr(nouvelle_pos + ord('A'))
        else:
            resultat += c                     # espaces inchangés
    return resultat

# On teste les 26 décalages
for cle in range(26):
    print(f"clé {cle:2d} : {dechiffre_cesar(message, cle)}")
```

## 10. Retour à Bézout : variante récursive de l'algorithme d'Euclide étendu

```python
def euclide_etendu_rec(a,b):
    if b == 0:
        return a, 1, 0
    g, u1, v1 = euclide_etendu_rec(b, a % b)
    return (g,v1, u1 - (a // b) * v1)
print(euclide_etendu_rec(240,46))  # (6, -1, 3) car 48*(-1) + 18*3 = 6
```

### 11.2 Matrices de rotation — lien avec les nombres complexes et les groupes

On vérifie numériquement puis symboliquement que $A(\theta)A(\theta') = A(\theta+\theta')$ et $A(\theta)^n = A(n\theta)$ : les matrices de rotation forment un groupe isomorphe à $(\mathbb{U}, \times)$ (le cercle unité de $\mathbb{C}$) via $\theta \mapsto e^{i\theta}$.

```python
import numpy as np
import sympy as sp

# --- Vérification numérique ---
def A(theta):
    return np.array([[np.cos(theta), -np.sin(theta)],
                     [np.sin(theta),  np.cos(theta)]])

theta, theta_p = 0.7, 1.2

# A(θ)·A(θ') == A(θ+θ') ?
print(np.allclose(A(theta) @ A(theta_p), A(theta + theta_p)))  # True

# A(θ)^n == A(nθ) ?
n = 5
print(np.allclose(np.linalg.matrix_power(A(theta), n), A(n*theta)))  # True

# --- Preuve symbolique avec sympy ---
th, thp = sp.symbols('theta theta_prime', real=True)

def As(t):
    return sp.Matrix([[sp.cos(t), -sp.sin(t)],
                      [sp.sin(t),  sp.cos(t)]])

produit = sp.simplify(sp.trigsimp(As(th) * As(thp)))
print(produit)          # Matrix([[cos(θ+θ'), -sin(θ+θ')], [sin(θ+θ'), cos(θ+θ')]])
print(As(th + thp))     # identique
```
